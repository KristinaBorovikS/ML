```python
import sqlite3
import pandas as pd
import seaborn as sns
import seaborn.objects as so
import matplotlib.pyplot as plt
from seaborn._core.plot import Plot

connection = sqlite3.connect('database_trees.db')

c = connection.cursor()

#c.execute( "DROP TABLE trees")

c.execute("""CREATE TABLE IF NOT EXISTS trees (
           latitude FLOAT,
           longitude  FLOAT,
           uuid  VARCHAR,
           gattung_botanisch VARCHAR,
           gattung_deutsch VARCHAR,
           art_botanisch VARCHAR,
           art_deutsch VARCHAR,
           alleebaum  FLOAT,
           hoehe FLOAT,
           stammdurchmesser FLOAT,
           stammumfang FLOAT,
           kronendurchmesser FLOAT,
           baumpatenschaft FLOAT
    )""")

df = pd.read_csv(r'baeume.csv')
print(df.head())
print(df.info())


(Plot.show(so.Plot(df.head(30), "gattung_botanisch", "kronendurchmesser")
.add(so.Dots(), so.Jitter())
.add(so.Range(),so.Shift(y=.5) )))

sns.swarmplot(data=df.head(150), x="hoehe", hue="gattung_botanisch",palette=sns.mpl_palette("viridis", 25),legend=False)
plt.show()
sns.boxenplot(data=df.head(170), x="latitude",palette=sns.mpl_palette("viridis", 25) ,legend=False)
plt.show()
sns.boxenplot(data=df.head(170), x="stammdurchmesser",palette=sns.light_palette("seagreen") ,legend=False)
plt.show()
sns.lmplot(data=df.head(20), x="hoehe", y="stammdurchmesser", hue="gattung_deutsch",col="gattung_botanisch")
plt.show()


df_heat= df.head(50).pivot(index="gattung_botanisch", columns="latitude", values="hoehe")
sns.heatmap(df_heat, cmap="crest")
plt.show()

df.to_sql('trees', connection, if_exists='append', index = False)



c.execute("""SELECT latitude,longitude,gattung_botanisch FROM trees
WHERE   hoehe = 5.00 AND stammdurchmesser= 6.37  AND kronendurchmesser =1
GROUP BY gattung_botanisch
HAVING COUNT(*) >= 20
ORDER BY gattung_deutsch
""")
for x in c.fetchall():
  print(x)



c.execute("""SELECT SUM(hoehe) as sum
FROM trees
WHERE gattung_botanisch ='Acer'
""")
print(c.fetchall())
c.execute("""SELECT gattung_botanisch,kronendurchmesser,stammdurchmesser FROM trees
GROUP BY gattung_botanisch
HAVING COUNT(*) >= 100
""")
for x in c.fetchall():
  print(x)

c.execute("""SELECT AVG(hoehe) as sum
FROM trees
WHERE gattung_botanisch ='Acer'
""")
print(c.fetchall())


c.execute("""SELECT COUNT(DISTINCT gattung_botanisch ) FROM trees
""")
print(c.fetchall())

connection.commit()
connection.close()
```
